```python
from langchain_text_splitters import HTMLHeaderTextSplitter
```

```python
html_string = """
<!DOCTYPE html>
<html>
  <body>
    <h1>Artificial Intelligence</h1>
    <p>Artificial Intelligence is a broad field of computer science.</p>

    <h2>Machine Learning</h2>
    <p>Machine learning allows systems to learn from data.</p>

    <h3>Supervised Learning</h3>
    <p>Supervised learning uses labeled data for training.</p>

    <h3>Unsupervised Learning</h3>
    <p>Unsupervised learning works with unlabeled data.</p>

    <h2>Deep Learning</h2>
    <p>Deep learning uses neural networks with many layers.</p>
  </body>
</html>
"""
```

```python
print(html_string)
```

```python
headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]
```

```python
print(headers_to_split_on)
```

```python
html_splitter = HTMLHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on
)
```

```python
html_docs = html_splitter.split_text(html_string)
```

```python
print("Total chunks:", len(html_docs))
```

```python
print("First chunk content:\n")
print(html_docs[0].page_content)

print("\nFirst chunk metadata:\n")
print(html_docs[0].metadata)
```

```python
for i, doc in enumerate(html_docs):
    print(f"\n--- Chunk {i+1} ---")
    print("Content:")
    print(doc.page_content)
    print("Metadata:")
    print(doc.metadata)
```

```python
for i, doc in enumerate(html_docs):
    print(f"Chunk {i+1}: {doc.page_content}")
```

```python
for i, doc in enumerate(html_docs):
    print(f"Chunk {i+1} metadata: {doc.metadata}")
```

```python
html_splitter_each = HTMLHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on,
    return_each_element=True
)
```

```python
html_docs_each = html_splitter_each.split_text(html_string)
```

```python
print("Total element-wise chunks:", len(html_docs_each))
```

```python
for i, doc in enumerate(html_docs_each):
    print(f"\n--- Element Chunk {i+1} ---")
    print("Content:")
    print(doc.page_content)
    print("Metadata:")
    print(doc.metadata)
```
